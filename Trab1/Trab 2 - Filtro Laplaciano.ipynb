Convolução 2D - Filtragem de Imagem
Filtro passa-baixa: remove ruídos ou borra a imagem
Filtro passa-alta: ajuda a encontrar bordas na imagem

### A imagem a ser utilizada: lenna.jpg

```python
img = mpimg.imread('lenna.jpg') # read image from file to array
imgplot = plt.imshow(img, cmap='gray')
```

### A função GaussianBlur 
Esta função aplica um filtro de suaviazação Gaussiano na imagem de entrada. Ela é fornecida na biblioteca OpenCV do python 3 e seus parâmetros são como segue: <br>
dst = cv2.GaussianBlur(src, ksize, sigmaX[, dst[, sigmaY[, borderType=BORDER_DEFAULT]]])<br>
src: imagem de entrada<br>
dst: imagem de saída<br>
ksize: Tamanho do kernel Gaussiano, (altura, largura). Ambos devem ser ímpares e podem ter diferentes valores. Se for definido [0 0], o ksize será calculado do valor de sigma<br>
sigmaX/Y: valor do desvio padrão no eixo X/Y;<br><br>
No exemplo a seguir temos um filtro com mascara de 5 x 5 e 15 x 15, com desvios padrão em x e y iguais a 10.

```python
gaussFilter = cv2.GaussianBlur( img, (5,5), 10, 10, 0)
gaussFilter2 = cv2.GaussianBlur( img, (15,15), 10, 10, 0)
plt.figure(figsize=(15,15))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.subplot(1,3,2)
plt.imshow(gaussFilter, cmap = 'gray');
plt.subplot(1,3,3)
plt.imshow(gaussFilter2, cmap = 'gray');
```

```python
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-10,10, num=100)
y=np.linspace(-10,10, num=100)

x, y = np.meshgrid(x, y)

z = np.exp(-0.1*x**2-0.1*y**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z, cmap=cm.jet)
plt.show()
```
